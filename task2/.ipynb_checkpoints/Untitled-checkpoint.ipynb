```python
idx = []
count_playes = len(pdf)
count_anwer = 0
count_anwers_team = 0
y = []
for id_tour in tournaments_train['id']:
    for team in results[id_tour]:
        list_players = []
        for player in team['teamMembers']:
            list_players.append(pdf['index'][player['player']['id']])
        if not 'mask' in team.keys() or team['mask'] is None:
            continue
        for anwer in team['mask']:
            if anwer == '1' or anwer == '0':
                for id_player in list_players:
                    idx.append([ count_anwer, id_player])
                    idx.append([count_anwer, count_playes +  count_anwers_team ])
                    count_anwer += 1
                    y.append(int(anwer))
            count_anwers_team += 1   
```

```python
assert len(y) == len(idx)/ 2
```

```python
X = sparse.lil_matrix((len(y), count_playes +  count_anwers_team))
```

```python
for i, j in idx:
    X[i, j] = 1
```

```python
y = np.array(y)
dim0, dim1 = X.shape
```

```python
lr = LogisticRegression(tol=1e-1, solver="saga", C=10, fit_intercept=False)
lr.fit(X, y)
```

```python
rating = lr.coef_[0][:count_playes]
```

```python
corr_1 = []
corr = []
for id_tour in tournaments_test['id']:
# for id_tour in [4957]:
    count_anw = []
    list_index_players = []
    j = 0
    if results[id_tour] == []:
        continue
   
    for team in results[id_tour]:
        if not 'mask' in team.keys() or team['mask'] is None:
            continue
        count_anw.append(len(re.findall(r'1',team['mask'])))
        list_append = []
        for playes in team['teamMembers']:
            list_append.append(pdf['index'][playes['player']['id']])
        list_index_players.append(list_append)
    if len(list_index_players) < 3:
        continue
    X = sparse.lil_matrix((len(list_index_players), dim1))
    for i in range(len(list_index_players)):
        for j in list_index_players[i]:
            X[i,j] = 1
    
    a = lr.predict_proba(X)[:, 1]
    corr_1.append(sts.spearmanr(a, count_anw).correlation)
    
    corr.append(sts.kendalltau(a, count_anw).correlation)
    
    
  
```
